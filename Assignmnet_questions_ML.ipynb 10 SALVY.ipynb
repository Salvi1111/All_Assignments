### What is the distribution  of preferred foot?


```python
mode1= df["Preferred Foot"].mode().values[0]
mode1
```

```python
df["Preferred Foot"].replace(np.nan, mode1, inplace=True)
```

```python
sns.countplot(x="Preferred Foot", data=df)
plt.show()
```

### How can we visualize the age distribution of players?

```python
plt.figure(figsize=(15,12))

sns.countplot(x="Age", data=df)
```

### What is the international reputation of left and right foot?

```python
mean1= df["International Reputation"].mean()
df["International Reputation"].replace(np.nan, mean1, inplace=True)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="International Reputation", hue="Preferred Foot", data=df)
```

### What is the different work rate of the Players?

```python
mode2= df["Work Rate"].mode().values[0]
df["Work Rate"].replace(np.nan, mode2, inplace=True)
```

```python
plt.figure(figsize=(15,8))

sns.countplot(x="Work Rate", data=df)
plt.show()
```

### Top 10 countries with most number of players

```python
mode3= df["Nationality"].mode().values[0]
df["Nationality"].replace(np.nan, mode3, inplace=True)
```

```python
A=df["Nationality"].value_counts(ascending=False)
A.head(10).plot(kind="bar")
plt.show()
```

### What is the work rate of players segregated by preferred foot?

```python
plt.figure(figsize=(15,10))

sns.countplot(x="Work Rate", hue="Preferred Foot", data=df)
plt.show()
```

### Number of players at different positions? 



```python
mode5= df["Position"].mode().values[0]
df["Position"].replace(np.nan, mode5, inplace= True)
```

```python
df["Position"].value_counts()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x="Position", data= df, palette= "Set2")
plt.xlabel= "Position"
plt.ylabel="No.of Players"
plt.title= "Number of players at different positions"
plt.show()
```

### Age distribution in top countries

```python
A= df['Nationality'].value_counts()[:10].index
print(A)

plt.figure(figsize=(15,12))

B = df.loc[df['Nationality'].isin(A)]
plt.figure(1 , figsize = (10,6))
sns.boxplot(x = 'Nationality' , y = 'Age' , data =B)

plt.show()
```
