```python
# Create a plot to check the relationship between horsepower and acceleration. Note down your insight for the same. Beautify the graph using various customizations.

plt.figure(figsize=(8, 6))
plt.scatter(mpg_df['horsepower'], mpg_df['acceleration'], color='blue', alpha=0.7)
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.title('Relationship between Horsepower and Acceleration')

plt.grid(True, linestyle='--', alpha=0.5)

plt.colorbar(label='Data Density')
plt.show()
```

```python
# Generate subplots to display the histograms for acceleration, displacement and weight and kilometer_per_litre. Note down your insight for the same
mpg_df["kmpl"]= mpg_df["mpg"]*1.609
mpg_df
```

```python
sns.histplot(x="acceleration", data=mpg_df)
plt.subplot()
plt.show()

sns.histplot(x="displacement", data=mpg_df, color="red")
plt.subplot()
plt.show()

sns.histplot(x="weight", data=mpg_df, color="yellow")
plt.subplot()
plt.show()

sns.histplot(x="kmpl", data=mpg_df, color="purple")
plt.subplot()
plt.show()
```

```python
#Create a plot to check the relationship between displacement and weight for origin 2, color the datapoints based on no of cylinders. Note down your insight for the same and save the plot as an image file.
```

```python
sns.relplot(x="displacement", y="weight", hue="cylinders", data=mpg_df)
plt.plot()
```

```python
# Display the frequency distribution of the Origin variable. Note down your insight for the same

sns.countplot(x= "origin", data= mpg_df)
plt.show()
```

```python
# Display the average weight based on no of cylinders present. Note down the insight for the same

avg_weight_by_cylinders = mpg_df.groupby('cylinders')['weight'].mean()

avg_weight_by_cylinders.plot(kind='bar')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average Weight')
plt.title('Average Weight based on Number of Cylinders')

plt.show()
```

```python
#  Check for the outliers in the dataset

sns.boxplot(data=mpg_df)
plt.title('Box Plot of the Dataset')
plt.show()
```

```python
# Plot the correlations for variables.

correlation_matrix = mpg_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
